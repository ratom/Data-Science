```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dropout,Dense,BatchNormalization,Flatten,Conv1D,MaxPool1D
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold
from tensorflow.keras.optimizers import Adam
```

```python
data=pd.read_csv('train.csv')
data.head()
```

```python
data.shape
```

```python
X=data.drop(labels=['ID','TARGET'], axis=1)
y=data['TARGET']
```

```python
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0,stratify=y)#prepared for training and testing data set
```

```python
X_train.shape,X_test.shape #to check the size of training datasets
```

```python
#remove quaisi constants, constants and duplicate features
filter=VarianceThreshold(0.1)
X_train=filter.fit_transform(X_train)
X_test=filter.transform(X_test)
print("hello")
```

```python
X_train.shape,X_test.shape
```

```python
#to transpose the columns and remove the duplicated columns if presented
X_train_T=X_train.T
X_test_T=X_test.T
#to convert transposed data into dataframe
X_train_T=pd.DataFrame(X_train_T)
X_test_T=pd.DataFrame(X_test_T)
X_train_T.shape,X_test_T.shape
```

```python
print("duplicated list****")
X_train_T.duplicated().sum()
print("duplicated list")
```

```python
X_train=X_train_T.T
X_train.shape
```

```python
X_test=X_test_T.T
X_test.shape
```

```python
X_train
```

```python
#the scale of X_train data is very differet.
#we do the standarise of the data to reduce chance of overfitting
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.shape,X_test.shape
```

```python
#the result of training set will be numpy array. So chage y to numpy. y is a series now
y_train=pd.DataFrame(y_train).to_numpy()
y_test=pd.DataFrame(y_test).to_numpy()
```

```python
#we need 3 D data
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1],1)
```

```python
X_train.shape,X_test.shape
```

```python
#building CNN
input_shape=X_train[0].shape
input_shape
```

```python
epochs=10
model=Sequential()

model.add(Conv1D(32,2, activation='relu',input_shape=X_train[0].shape))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Conv1D(64,2, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(lr=0.0001),loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=epochs, validation_data=(X_test,y_test),verbose=1)
```
